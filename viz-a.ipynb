# Data Visualization 

## Data Set : ["Car Crash"]('../input/car-crash-dataset')


```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_crashes = sns.load_dataset('car_crashes')
```

```python
sns.set()
car_crashes.head(5)
```

 ### - Visulize our data as total crashes cares to for every state :


```python
sns.relplot(y='total', x='abbrev', data=car_crashes, height=15);
```

### by other way 

```python
sns.barplot(y='total', x='abbrev', data=car_crashes);
```

### - Total Crashes by speeding 

```python
car_crashes_cile = np.ceil(car_crashes.loc[:, car_crashes.columns != 'abbrev'])
by_speeding = car_crashes_cile.groupby('speeding', as_index=False).sum()

sns.barplot(x='total', y='speeding', data=by_speeding);
```
