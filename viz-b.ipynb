```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('taxis')
df.head()
```

```python
taxis = df[df.passengers== 2]
pickup_borough = taxis.groupby("pickup_borough", as_index=False).tip.mean()
sns.barplot(x="pickup_borough",y="tip",data=pickup_borough)
pickup_borough
```

```python
avg_mpg = df.groupby("pickup_borough", as_index=False).tip.mean()
sns.relplot(x="pickup_borough",y="tip",data=avg_mpg)
```
